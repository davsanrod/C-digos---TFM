```python
#Afinamos la busqueda de los enlaces
enlaces = bSoup.find(id="ir_1254736046388").find_all('option')
enlaces
```

```python
#Guardamos los enlaces en una lista
descargaDatos = []
for i in bSoup.find(id="ir_1254736046388").find_all('option'):
    descargaDatos.append(i.get('value'))
```
